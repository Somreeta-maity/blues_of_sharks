```python
total_amount_per_shark = {}
for col in  ['ashneer_deal', 'anupam_deal', 'aman_deal',
       'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']:
     total_amount_per_shark[col] = [np.sum(win[win[col] == 1].loc[:,'amount_per_shark'])]

total_equity_per_shark = {}
for col in [ 'ashneer_deal', 'anupam_deal', 'aman_deal',
       'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']:
    total_equity_per_shark[col] = [np.sum(win[win[col] == 1].loc[:,'equity_per_shark'])]
    
plt.figure(figsize = (15,4))
plt.subplot(1,2,1)
sns.barplot(x = 'index',y = 'Total Amount' ,data =  pd.DataFrame(total_amount_per_shark).T.reset_index().rename(columns = {0:'Total Amount'}))
plt.xticks(rotation = 90)
plt.subplot(1,2,2)
sns.barplot(x = 'index',y = 'Total Equity' ,data =  pd.DataFrame(total_equity_per_shark).T.reset_index().rename(columns = {0:'Total Equity'}))
plt.xticks(rotation = 90)
plt.show()
```

```python
sharks = ['ashneer_deal', 'anupam_deal', 'aman_deal',
       'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']
mat = []
for i in range(len(sharks)):
    temp = []
    for j in range(len(sharks)):
        x = sharks[i]
        y = sharks[j]
        temp.append((win.query(f"{x} == 1 and {y} == 1").shape[0]/win.query(f"{x} == 1 or {y} == 1").shape[0])*100)
    mat.append(temp)
    
shark_df = pd.DataFrame(mat, columns = sharks, index = sharks)

plt.figure(figsize = (12,10))
sns.heatmap(data = shark_df,vmin = 0, vmax = 100, cmap = 'Blues', annot = True, fmt = 'g')
plt.show()
```
